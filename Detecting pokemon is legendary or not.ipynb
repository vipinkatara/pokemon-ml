# Dropping the '#' column

```python
df_pair = df.drop('#', axis=1)
```

# Visualizing the data

```python
sns.pairplot(df_pair, hue='Legendary', height=10)
plt.show()
```

# Initializing the standard scaler z = x - u/var

```python
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
```

# Label Encoder

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
df_pair['Name'] = le.fit_transform(df_pair['Name'])
df_pair['Type 1'] = le.fit_transform(df_pair['Type 1'])
```

```python
df_pair['Type 2']=df_pair['Type 2'].fillna('no ability')
df_pair['Type 2'] = le.fit_transform(df_pair['Type 2'])
```

```python
df_pair
```

```python
df_pair
```

# separating labels and features

```python
label = df_pair['Legendary']
df_pair = df_pair.drop('Legendary', axis=1)
data = ss.fit_transform(df_pair)
```

# Applying different dimensionaltiy reduction technique(PCA and TSNE)

```python
from sklearn.manifold import TSNE
from sklearn import decomposition
```

```python
pca = decomposition.PCA()
pca.n_components=2
pca_data = pca.fit_transform(data)
pca_data = pca_data.T
pca_data = np.vstack((pca_data, label)).T
pca_df = pd.DataFrame(data=pca_data, columns=('dim1', 'dim2', 'labels'))
sns.FacetGrid(pca_df, hue='labels', height=10).map(plt.scatter, 'dim1', 'dim2').add_legend()
```

```python
model = TSNE(n_components=2, random_state=0, perplexity=30, n_iter=1000)
tsne_data = model.fit_transform(data)

tsne_data = np.vstack((tsne_data.T, label)).T
df_tsne = pd.DataFrame(data= tsne_data, columns=('dim1', 'dim2', 'label'))

sns.FacetGrid(df_tsne, hue='label', height=10).map(plt.scatter, 'dim1', 'dim2').add_legend()
```

```python
model = TSNE(n_components=2, random_state=0, perplexity=30, n_iter=5000)
tsne_data = model.fit_transform(data)

tsne_data = np.vstack((tsne_data.T, label)).T
df_tsne = pd.DataFrame(data= tsne_data, columns=('dim1', 'dim2', 'label'))

sns.FacetGrid(df_tsne, hue='label', height=10).map(plt.scatter, 'dim1', 'dim2').add_legend()
```

```python
model = TSNE(n_components=2, random_state=0, perplexity=30, n_iter=1000)
tsne_data = model.fit_transform(data)

tsne_data = np.vstack((tsne_data.T, label)).T
df_tsne = pd.DataFrame(data= tsne_data, columns=('dim1', 'dim2', 'label'))

sns.FacetGrid(df_tsne, hue='label', height=10).map(plt.scatter, 'dim1', 'dim2').add_legend()
```

```python
model = TSNE(n_components=2, random_state=0, perplexity=50, n_iter=1000)
tsne_data = model.fit_transform(data)

tsne_data = np.vstack((tsne_data.T, label)).T
df_tsne = pd.DataFrame(data= tsne_data, columns=('dim1', 'dim2', 'label'))

sns.FacetGrid(df_tsne, hue='label', height=10).map(plt.scatter, 'dim1', 'dim2').add_legend()
```

# Spliting the dataset into train and test

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(df_pair, label, test_size=0.2, random_state=0)
```

# Applying Multinomial classification on training data and predicting its accuracy 

```python
from sklearn.naive_bayes import MultinomialNB
classifier = MultinomialNB()
classifier.fit(x_train, y_train)
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
pred = classifier.predict(x_test)
print(classification_report(y_test, pred))
print()
print('Confusion Matrix:\n',confusion_matrix(y_test, pred))
print()
print('Accuracy : ',accuracy_score(y_test, pred))
```

# Applying SVM classification on training data and predicting its accuracy 

```python
from sklearn import svm
svm = svm.SVC()
svm.fit(x_train, y_train)
```

```python
pred = svm.predict(x_test)
print(classification_report(y_test, pred))
print()
print('Confusion Matrix:\n',confusion_matrix(y_test, pred))
print()
print('Accuracy : ',accuracy_score(y_test, pred))
```

#  Applying KNN classification on training data and predicting its accuracy 

```python
from sklearn.neighbors import KNeighborsClassifier
neigh = KNeighborsClassifier(n_neighbors=3)
neigh.fit(x_train, y_train)
```

```python
pred = neigh.predict(x_test)
print(classification_report(y_test, pred))
print()
print('Confusion Matrix:\n',confusion_matrix(y_test, pred))
print()
print('Accuracy : ',accuracy_score(y_test, pred))
```

# Applying Decision tree classification on training data and predicting its accuracy 

```python
from sklearn import tree
clf_tree = tree.DecisionTreeClassifier()
clf_tree.fit(x_train,y_train)
```

```python
pred = clf_tree.predict(x_test)
print(classification_report(y_test, pred))
print()
print('Confusion Matrix:\n',confusion_matrix(y_test, pred))
print()
print('Accuracy : ',accuracy_score(y_test, pred))
```

# Applying Random Forest classification on training data and predicting its accuracy 

```python
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier(max_depth=2, random_state=0)
clf.fit(x_train, y_train)
```

```python
pred = clf.predict(x_test)
print(classification_report(y_test, pred))
print()
print('Confusion Matrix:\n',confusion_matrix(y_test, pred))
print()
print('Accuracy : ',accuracy_score(y_test, pred))
```
